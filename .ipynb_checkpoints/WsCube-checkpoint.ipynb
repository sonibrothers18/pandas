# Merge

```python
df1 = pd.DataFrame({
    "A": [1,2,3,4],
    "B": [11,12,13,14]
})

# Should have one common part

df2 = pd.DataFrame({
    "A": [1,2,3,4],
    "C": [21,22,23,24]
})
```

```python
pd.merge(df2, df1, on="A")
```

## Different values in common part -->

```python
df1 = pd.DataFrame({
    "A": [1,2,3,4],
    "B": [11,12,13,14]
})

df2 = pd.DataFrame({
    "A": [1,2,3,5],
    "C": [21,22,23,24]
})

pd.merge(df1, df2, on="A")
```

```python
pd.merge(df1, df2, how="left") # like left joint
```

```python
pd.merge(df1, df2, how="outer")
```

```python
pd.merge(df1, df2, how="outer", indicator=True)
```

## Same column name -->

```python
df1 = pd.DataFrame({
    "A": [1,2,3,4],
    "B": [11,12,13,14]
})

df2 = pd.DataFrame({
    "A": [1,2,3,5],
    "B": [21,22,23,24]
})

pd.merge(df1, df2, left_index=True, right_index=True)
```
